```python
def printDiagonal(lst):
    # To print Primary Diagonal
    print('Diagonal 1 -'),
    print([lst[i][i] for i in range(len(lst))])
 
    # To print Secondary Diagonal
    print('Diagonal 2 -'),
    print([lst[i][len(lst)-1-i] for i in range(len(lst))])
 
     
# Driver code
lst = [[1, 2, 3],
       [4, 5, 6],
       [7, 8, 9]]
 
printDiagonal(lst)
```
