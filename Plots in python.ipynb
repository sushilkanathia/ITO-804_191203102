```python
import matplotlib.pyplot as plt
```

```python
#line plot
plt.plot([1,2,3,5,6], [1, 2, 3, 4, 6] , color = 'magenta', linestyle = "dashed",marker = 'o',
         markerfacecolor = 'red', markersize = 12) #coordinates of points -> (1,1) (2,2) (3,3) (5,4) (6,6)
plt.plot([2,3,4,5,6], [1, 2, 3, 4, 5] , color = 'green')
#plt.axis([0, 7, 0, 10])  #size of plotting area :x -> 0-7 ; y-> 0-10

# Print the chart
plt.show()
```

```python
#scatter plot

x1 = [1, 2,3,4,5,6,7]
y1 = [1,2, 3, 1, 2, 3, 4]
pyplot.scatter(x1, y1, label = 'high bp low heartrate', color='c')
pyplot.title('Smart Band Data Report')
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.legend()

# Print the chart
pyplot.show()
```

```python
#histogram plot
pop = [5,10,15,20]
bins = [1,4,7,11]
pyplot.hist(pop, bins, rwidth=0.1)
pyplot.xlabel('pop')
pyplot.ylabel('bin')
pyplot.title('Histogram')

# Print the chart
pyplot.show()
```

```python
#area plot 
days = [1,2,3,4,5]
age =[63, 81, 52, 22, 37]
weight =[17, 28, 72, 52, 32]
pyplot.plot([],[], color='c', label = 'Weather Predicted', linewidth=5)
pyplot.plot([],[],color = 'g', label='Weather Change happened', linewidth=5)
pyplot.stackplot(days, age, weight, colors = ['c', 'g'])
pyplot.xlabel('Fluctuation with time')
pyplot.ylabel('Days')
pyplot.title('Weather report using Area Plot')
pyplot.legend()

# Print the chart
pyplot.show()
```
